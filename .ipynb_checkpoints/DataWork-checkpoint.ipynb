```python
#Average out weight
team_mean = data.groupby('team')['weight(kg)'].transform('mean')
nat_mean = data.groupby('nationality')['weight(kg)'].transform('mean')
pos_mean = data.groupby('field_position')['weight(kg)'].transform('mean')
data['height_bin'] = pd.cut(data['height(cm)'], bins=5)
height_mean = data.groupby('height_bin', observed=True)['weight(kg)'].transform('mean')
combined = (
    0.25 * team_mean +
    0.25 * nat_mean +
    0.25 * pos_mean +
    0.25 * height_mean
)

data['weight(kg)'] = data['weight(kg)'].fillna(combined)
data['weight(kg)'] = data['weight(kg)'].fillna(data['weight(kg)'].mean())
data['weight(kg)'] += np.random.normal(0, 1.5, size=len(data))
```

```python
#Now I will deal with height:
import numpy as np

# Track missing
missing_mask = data['height(cm)'].isna()

# Group means for HEIGHT (not weight)
nat_h = data.groupby('nationality')['height(cm)'].transform('mean')
pos_h = data.groupby('field_position')['height(cm)'].transform('mean')

# Combine (position matters more)
combined_h = (
    0.6 * pos_h +
    0.4 * nat_h
)

# Fill missing
data.loc[missing_mask, 'height(cm)'] = combined_h[missing_mask]

# Fallback
data['height(cm)'] = data['height(cm)'].fillna(data['height(cm)'].mean())

# Add noise ONLY to imputed values
data.loc[missing_mask, 'height(cm)'] += np.random.normal(0, 1.5, size=missing_mask.sum())
```

### Feature Engineering

```python
data.sample()
```

```python
data['avg_goals_per_game'] = data['goals']/(data['minutes_played']/90)
data['avg_goals_per_game'] = data['avg_goals_per_game'].fillna(0)
```

```python
data.sample()
```

```python
data['assists_per_90'] = data['assists']/(data['minutes_played']/90)
data['assists_per_90'] = data['assists_per_90'].fillna(0)
```

```python
data.sample()
```

```python
data['passes_per_90'] = data['passes_completed'] / (data['minutes_played'] / 90)
data['team_index'] = (
    0.4 * data['passing_accuracy(%)'] +
    0.3 * data['passes_per_90'] + 0.1 * data['passes_attempted'] + 0.2 * data['assists_per_90']
)
data['passes_per_90'] = data['passes_per_90'].fillna(0)
data['team_index'] = data['team_index'].fillna(0)
```

```python
data.sample()
```

```python
data['showman_index'] = (data['head'] + data['dribbles'] + data['red_cards'] + data['yellow_cards']) / (data['minutes_played'] / 90)
```

```python
data['appearance'] = data['minutes_played'] * 0.5 + data['matches_appearance'] * 0.5
```

```python
data['distance_covered(km/h)'].max()
```

```python
data['aggression_index'] = (
    1.0 * data['tackles_won'] +
    1.5 * data['fouls_committed'] +
    3.0 * data['yellow_cards'] +
    6.0 * data['red_cards']
) / (data['minutes_played'] / 90)
```

```python
data['runner_index'] = (
    data['distance_covered(km/h)'] +
    data['top_speed']
)
```

```python
data['saves_per_90']= data['saves'] / (data['minutes_played'] / 90)
data['clean_sheets_per_90'] = data['clean_sheets'] / (data['minutes_played'] / 90)
data['goals_conceded_per_90'] = data['goals_conceded'] / (data['minutes_played'] / 90)
data['punches_made_per_90'] = data['punches_made'] / (data['minutes_played'] / 90)
```

```python
data = data.fillna(0)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
import numpy as np

# =========================
# 1️⃣ Fit scalers and KNN models
# =========================

# Outfield features
outfield_features = [
    'height(cm)', 'weight(kg)', 'avg_goals_per_game',
    'team_index', 'showman_index', 'top_speed',
    'runner_index', 'aggression_index'
]
scaler_out = StandardScaler()
out_scaled = scaler_out.fit_transform(outfield_data[outfield_features])

knn_out = NearestNeighbors(n_neighbors=5)
knn_out.fit(out_scaled)

# Goalkeeper features
gk_features = [
    'saves_per_90', 'clean_sheets_per_90', 'goals_conceded_per_90',
    'punches_made_per_90', 'height(cm)', 'weight(kg)', 'saves_on_penalty'
]
scaler_gk = StandardScaler()
gk_scaled = scaler_gk.fit_transform(gk_data[gk_features])

knn_gk = NearestNeighbors(n_neighbors=5)
knn_gk.fit(gk_scaled)

# =========================
# 2️⃣ Recommendation function
# =========================
def recommend_player():
    position = input("Do you want a Goalkeeper or Outfielder? ").strip().lower()

    def clip_or_warn(val, col, data_col):
        min_val, max_val = data_col.min(), data_col.max()
        if val < min_val or val > max_val:
            print(f"⚠ Warning: {col} should be between {min_val:.2f} and {max_val:.2f}. Clipping value.")
            val = np.clip(val, min_val, max_val)
        return val

    # =========================
    # 🧤 GOALKEEPER SECTION
    # =========================
    if position == "goalkeeper":
        print("\nEnter your preferred stats:\n")
        saves = clip_or_warn(float(input("Saves per 90: ")), "Saves per 90", gk_data['saves_per_90'])
        clean = clip_or_warn(float(input("Clean sheets per 90: ")), "Clean sheets per 90", gk_data['clean_sheets_per_90'])
        conceded = clip_or_warn(float(input("Goals conceded per 90: ")), "Goals conceded per 90", gk_data['goals_conceded_per_90'])
        punches = clip_or_warn(float(input("Punches made per 90: ")), "Punches made per 90", gk_data['punches_made_per_90'])
        height = clip_or_warn(float(input("Height (cm): ")), "Height", gk_data['height(cm)'])
        weight = clip_or_warn(float(input("Weight (kg): ")), "Weight", gk_data['weight(kg)'])
        pens = clip_or_warn(float(input("Penalty saves (approx per 90): ")), "Penalty saves", gk_data['saves_on_penalty'])

        # Fix Warning & Transform
        user_input_df = pd.DataFrame([[saves, clean, conceded, punches, height, weight, pens]], 
                                     columns=gk_features)
        user_scaled = scaler_gk.transform(user_input_df)
        distances, indices = knn_gk.kneighbors(user_scaled)
        results = gk_data.iloc[indices[0]]

        print(f"\n🔥 Top 5 Recommended Goalkeepers (UCL 24/25 Data):")
        print("="*75)

        for _, row in results.iterrows():
            similarity_sum = 0
            knn_stats_breakdown = []
            
            for f in gk_features:
                target_val = user_input_df.iloc[0][f]
                player_val = row[f]
                min_f, max_f = gk_data[f].min(), gk_data[f].max()
                feat_range = max_f - min_f
                diff = abs(target_val - player_val) / feat_range if feat_range != 0 else 0
                similarity_sum += max(0, 1 - diff)
                knn_stats_breakdown.append(f"  - {f:22} | {row['player_name']}: {player_val:<7.2f} (Target: {target_val:.2f})")

            match_pct = (similarity_sum / len(gk_features)) * 100
            
            print(f"Name: {row['player_name']} ({row['team']})")
            print(f"Accuracy to your input: {match_pct:.1f}% | Archetype: {row['archetype_name']}")
            
            print("\n[SCALING & KNN STATS]")
            print("\n".join(knn_stats_breakdown))
            
            print("\n[PLAYER CONTEXT & UCL 2025 PERFORMANCE]")
            print(f"  - Age: {int(row['age'])}")
            print(f"  - Minutes Played in UCL 2025: {row['minutes_played']:,.0f}")
            print(f"  - Matches Appearance: {int(row['matches_appearance'])}")
            print(f"  - Passing Accuracy: {row['passing_accuracy(%)']:.1f}%")
            print(f"  - Balls Recovered: {row['balls_recovered']:.0f}")
            print("-" * 75)

    # =========================
    # 🏃 OUTFIELDER SECTION
    # =========================
    elif position == "outfielder":
        print("\nEnter your preferred playstyle:\n")
        height = clip_or_warn(float(input("Height (cm): ")), "Height", outfield_data['height(cm)'])
        weight = clip_or_warn(float(input("Weight (kg): ")), "Weight", outfield_data['weight(kg)'])
        goals = clip_or_warn(float(input("Goals per game: ")), "Goals per game", outfield_data['avg_goals_per_game'])
        team_idx = clip_or_warn(float(input("Team playstyle (30–70): ")), "Team index", outfield_data['team_index'])
        showman = clip_or_warn(float(input("Showman index (0–10): ")), "Showman index", outfield_data['showman_index'])
        speed = clip_or_warn(float(input("Top speed (km/h): ")), "Top speed", outfield_data['top_speed'])
        runner = clip_or_warn(float(input("Runner index (~20–60): ")), "Runner index", outfield_data['runner_index'])
        aggression = clip_or_warn(float(input("Aggression index (0–10): ")), "Aggression index", outfield_data['aggression_index'])

        user_input_df = pd.DataFrame([[height, weight, goals, team_idx, showman, speed, runner, aggression]], 
                                     columns=outfield_features)
        user_scaled = scaler_out.transform(user_input_df)
        distances, indices = knn_out.kneighbors(user_scaled)
        results = outfield_data.iloc[indices[0]]

        print(f"\n🔥 Top 5 Recommended Outfielders (UCL 24/25 Data):")
        print("="*75)

        for _, row in results.iterrows():
            similarity_sum = 0
            knn_stats_breakdown = []
            
            for f in outfield_features:
                target_val = user_input_df.iloc[0][f]
                player_val = row[f]
                min_f, max_f = outfield_data[f].min(), outfield_data[f].max()
                feat_range = max_f - min_f
                diff = abs(target_val - player_val) / feat_range if feat_range != 0 else 0
                similarity_sum += max(0, 1 - diff)
                knn_stats_breakdown.append(f"  - {f:22} | {row['player_name']}: {player_val:<7.2f} (Target: {target_val:.2f})")

            match_pct = (similarity_sum / len(outfield_features)) * 100
            
            print(f"Name: {row['player_name']} ({row['team']})")
            print(f"Match Score: {match_pct:.1f}% | Archetype: {row['archetype_name']}")
            
            print("\n[SCALING & KNN STATS]")
            print("\n".join(knn_stats_breakdown))
            
            print("\n[PLAYER CONTEXT & UCL 2025 PERFORMANCE]")
            print(f"  - Age: {int(row['age'])}")
            print(f"  - Minutes Played in UCL 2025: {row['minutes_played']:,.0f}")
            print(f"  - Passing Accuracy: {row['passing_accuracy(%)']:.1f}%")
            print(f"  - Dribbles (Total): {row['dribbles']:.0f}")
            print(f"  - Balls Recovered: {row['balls_recovered']:.0f}")
            print("-" * 75)

    else:
        print("❌ Invalid input. Please type 'Goalkeeper' or 'Outfielder'.")
```

```python
recommend_player()
```
